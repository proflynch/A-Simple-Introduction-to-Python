${\bf 1.}$ Investigate the differences between high- and low-income economies in the LDS about countries. 

```python
# 1. Compare death-rates per 1000 for each category, for example. 
import pandas as pd    # Import pandas for data analysis
import seaborn as sns  # Import seaborn for visualisations
import matplotlib.pyplot as plt
df1 = pd.read_excel("Data_1_OCR.xlsx" , sheet_name = "Data")
df1["Income category"] = (df1["GDP per capita (US$)"] > 12376).map({True: "High", False: "Low"})
# List the data frame with certain columns.
df1.iloc[0 : 5 , [0, 1, 2, 3, 4, 5, 6, 9, 20]] 
```

```python
# Create violin plots plots death-rate per 1000, grouped by income category.
sns.violinplot(data=df1, kind="box", x="death rate per 1000", y="Income category", aspect=2)
```

The lower/upper quartiles and median values are not that dissimilar. There are more outside points for the "Low Income" category. 

${\bf 2.}$ Compare the statistics for different features grouped by region or income categories: e.g. Use groupby("Region") and describe life expectancy at birth 2000.

```python
import pandas as pd    # Import pandas for data analysis
import seaborn as sns  # Import seaborn for visualisations
import matplotlib.pyplot as plt
df1 = pd.read_excel("Data_1_OCR.xlsx" , sheet_name = "Data")
df1["Income category"] = (df1["GDP per capita (US$)"] > 12376).map({True: "High", False: "Low"})
# List the data frame with certain columns.
df1.iloc[0 : 5 , [0, 1, 2,18,19, 20]] 
```

```python
# Create violin plots plots Life expectancy at birth 2000, grouped by income category.
sns.violinplot(data=df1, kind="box", x="Life expectancy at birth 2000", y="Income category", aspect=2)
```

These violin plots are markedly different. The overall "Life expectancy at birth 2000" is much worse for the "Low income category." There are more outside points for the "High Income" category.
